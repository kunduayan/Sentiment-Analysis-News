```python
import nltk
nltk.download('stopwords')
import pandas as pd
```

```python
def cleanText(text):
    """
    removes punctuation, stopwords and returns lowercase text in a list of single words
    """
    text = str(text).lower()    
    
    #from bs4 import BeautifulSoup
    #text = BeautifulSoup(text).get_text()
    
    from nltk.tokenize import RegexpTokenizer
    tokenizer = RegexpTokenizer(r'\w+')
    text = tokenizer.tokenize(text)
    
    from nltk.corpus import stopwords
    clean = [word for word in text if word not in stopwords.words('english')]
    
    return clean
```

```python
def loadPositive(path_postive):
    """
    loading positive dictionary
    """
    myfile = pd.read_excel(path_postive)
    positive=[]
    for i in range(myfile.shape[0]):
        positive.append(myfile.iloc[i,0].strip().lower())
    
    return positive
```

```python
def loadNegative(path_negative):
    """
    loading negative dictionary
    """
    myfile = pd.read_excel(path_negative)
    negative=[]
    for i in range(myfile.shape[0]):
        negative.append(myfile.iloc[i,0].strip().lower())
    
    return negative
```

```python
def countNeg(cleantext, negative):
    """
    counts negative words in cleantext
    """
    negs = [word for word in cleantext if word in negative]
    return len(negs)

def countPos(cleantext, positive):
    """
    counts negative words in cleantext
    """
    pos = [word for word in cleantext if word in positive]
    return len(pos)  
```

```python
def getSentiment(cleantext,positive,negative):
    """
    counts negative and positive words in cleantext and returns a score accordingly
    """
    x = countPos(cleantext, positive)
    y = countNeg(cleantext, negative)
    z = x-y
    if -1 <= z <= 1:
        return 0
    elif 2<= z :
        return 1
    elif -2 >= z:
        return -1
```

```python
def updateSentimentDataFrame(df):
    """
    performs sentiment analysis on single text entry of dataframe and returns dataframe with scores
    """
    path_postive='C:/Users/AYAN/Downloads/archive (1)/LM_Positive.xlsx'
    path_negative='C:/Users/AYAN/Downloads/archive (1)/LM_Negative.xlsx'
    positive = loadPositive(path_postive)
    negative = loadNegative(path_negative)
    
    df['content'] = df['content'].apply(cleanText)
    print('aa')
    df['score'] = df['content'].apply(lambda x: getSentiment(x,positive,negative))
    return df    
```
